The geometry is generated by a python script [`gen_geo.py`](https://github.com/czczc/RAT-analysis/blob/master/bnl1ton/gen_geo.py) and output to the RAT `.geo` file. All dimensions are taken from [designed or measure values](http://www.phy.bnl.gov/bnlif/docdb/0000/000096/012/TO%20DO%20One%20ton%20black%20Teflon%20optical%20barrier%20v15.pdf). Geometry can be easily modified in the python script.

The geometry includes the following main components:
- A dark box
- A central cylindrical acrylic detector
    - acrylic wall
    - cylindrical teflon sheets inside the wall
    - acrylic bottom
    - arrylic lid
    - center port on the lid
    - side port on the lid
- Six R7723 2'' PMTs ([implemetation](pmt.ipynb))
- Aluminum supporting platform and frames
- (to be added) muonn paddles

(to be added) All material definition.

Liqid (water or WbLS) is filled up to half of the port height.
